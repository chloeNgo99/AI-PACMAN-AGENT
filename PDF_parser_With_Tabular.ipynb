```python
from tabula import read_pdf
import pandas as pd
import warnings
import plotly.express as px
from datetime import datetime
```

```python
# scource url https://cdn.nar.realtor/sites/default/files/documents/phs-05-2023-pending-home-sales-06-29-2023.pdf
df = read_pdf("pending-home-sales-06-2023.pdf", pages='all')
df[0]
```

```python
#rename columns title
pending_home_sales = df[0].iloc[1:]
col_name = df[0].iloc[0]
pending_home_sales.columns = col_name
pending_home_sales.reset_index(drop=True, inplace=True)

pending_home_sales
```

```python
#remove nan columns
pending_home_sales = pending_home_sales.dropna(how='all', axis=1)
pending_home_sales
```

```python
#create the Average pending home sales through 2020-2022
pending_home_sales_avg = pending_home_sales.iloc[:3, :5]
# Split the 'Northeast Midwest' column on the space character and expand the results into separate columns
pending_home_sales_avg[['Northeast', 'Midwest']] = pending_home_sales_avg['Northeast Midwest'].str.split(' ', expand=True)

# Drop the original 'Northeast Midwest' column
pending_home_sales_avg = pending_home_sales_avg.drop(columns='Northeast Midwest')

pending_home_sales_avg
```

```python
#create a reusable function to extract data
def extract_data(new_table):
  # Split the 'Northeast Midwest' column on the space character and expand the results into separate columns
  new_table[['Northeast', 'Midwest']] = new_table['Northeast Midwest'].str.split(' ', expand=True)
  # Drop the original 'Northeast Midwest' column
  new_table = new_table.drop(columns='Northeast Midwest')

  # Convert Year to datetime
  # Remove trailing ' r' and ' p'
  new_table['Year'] = new_table['Year'].str.replace(' r', '').str.replace(' p', '')
  # Convert to datetime
  new_table['Year'] = pd.to_datetime(new_table['Year'])

  #reset row
  new_table.reset_index(drop=True, inplace=True)
  return new_table
```

```python
#create a Seasonally Adjusted Annual Rate table
table = pending_home_sales.iloc[4:17, :5]
seasonally_adjusted_annual_rate = extract_data(table)
seasonally_adjusted_annual_rate
```

```python
#create a Not Seasonally Adjusted table
table = pending_home_sales.iloc[4:17, 5:]
table['Year'] = pending_home_sales['Year'].iloc[4:17]

# Move the year columns to the front
table = pd.concat([table['Year'], table.drop('Year', axis=1)], axis=1)
not_seasonally_adjusted = extract_data(table)
not_seasonally_adjusted
```

```python
#create time plot for pending_home_sales_avg
fig = px.line(pending_home_sales_avg, x="Year", y=pending_home_sales_avg.columns[1:],
              title='Pending Home Sales Avg')
fig.show()
```

```python
#create time plot for Seasonally Adjusted Annual Rate table
fig = px.line(seasonally_adjusted_annual_rate, x="Year", y=seasonally_adjusted_annual_rate.columns[1:],
              title='Not Seasonally Adjusted Home Sales')
fig.show()
```

```python
#create time plot for Not Seasonally Adjusted table
fig = px.line(not_seasonally_adjusted, x="Year", y=not_seasonally_adjusted.columns[1:],
              title='Not Seasonally Adjusted Home Sales')
fig.show()
```

```python
#export tables to csv
pending_home_sales_avg.to_csv('pending_home_sales_avg.csv', index=False)
not_seasonally_adjusted.to_csv('not_seasonally_adjusted.csv', index=False)
seasonally_adjusted_annual_rate.to_csv('seasonally_adjusted_annual_rate.csv', index=False)
```
